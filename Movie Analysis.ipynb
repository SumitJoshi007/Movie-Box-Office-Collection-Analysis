```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("moviestats.csv")
df.head()
```

```python
# Define a function to handle case statement
def select_col(row):
    if row["MovieName"] == "RRR":
        return row["APTG"]
    elif row["MovieName"] == "Salaar":
        return row["APTG"]
    elif row["MovieName"] == "Baahubali 2":
        return row["APTG"]
    elif row["MovieName"] == "Goat":
        return row["TamilNadu"]
    elif row["MovieName"] == "Leo":
        return row["TamilNadu"]
    elif row["MovieName"] == "Jailer":
        return row["TamilNadu"]
    else:
        return row["RestOfIndia"]
    
df["homecollection"] = df.apply(select_col, axis=1)
  
```

```python
df
```

```python
#rajamouli difference

movie_day_1 = df[df["Day_Number"]==1]
movie_day_1.sort_values(by="Total", ascending=False)
```

```python
prabhas_data = df[(df["MovieName"] == "Baahubali 2") | (df["MovieName"] == "Salaar")]
prabhas_data
```

```python
#Restofindia impact

df["percent_roi"] = df["RestOfIndia"]/df["Total"]
df.sort_values(by="percent_roi", ascending=False)
```

```python
df["previousdaytotal"] = df.groupby("MovieName")["Total"].shift(1)
df["percent_chng"] = (df["Total"] - df["previousdaytotal"]) / df["previousdaytotal"]
df
```

```python
#home total

homedata = df.groupby("MovieName")["homecollection"].sum().reset_index()
homedata["rank"] = homedata["homecollection"].rank(method = 'dense', ascending=False)
homedata.sort_values(by='rank')
```

```python
#vijay vs rajni

leojailerdata = df[(df["MovieName"] == "Jailer") | (df["MovieName"] == "Leo")]
leojailerdata.groupby("MovieName")["Total"].sum().reset_index()
```
